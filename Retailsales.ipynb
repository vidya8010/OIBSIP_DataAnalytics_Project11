```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("retail_sales_dataset.csv")
df.head()
```

```python
df.info
```

```python
df.describe()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.fillna(0,inplace=True)
```

```python
df
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['Month'] = df['Date'].dt.month
df['Year']=df['Date'].dt.year
df['Day']=df['Date'].dt.day
```

```python
category_sales = df.groupby('Product Category')['Total Amount'].sum().reset_index()
sns.barplot(x='Product Category', y='Total Amount', data=category_sales)
plt.title('Total Sales by Product Category')
plt.show()
```

```python
gender_sales = df.groupby('Gender')['Total Amount'].sum().reset_index()
sns.barplot(x='Gender', y='Total Amount', data=gender_sales)
plt.title('Total Sales by Gender')
plt.show()
```

```python
top_customers = df.groupby('Customer ID')['Total Amount'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
top_customers.plot(kind='barh', color='skyblue')
plt.xlabel('Total Amount')
plt.ylabel('Customer ID')
plt.title('Top 10 Customers by Total Spending')
plt.gca().invert_yaxis()
plt.show()
print("\nTop 10 Customers:\n", top_customers)
```

```python
monthly_sales = df.groupby(df['Date'].dt.to_period('M'))['Total Amount'].sum()
plt.figure(figsize=(8,5))
monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.show()
print("\nMonthly Sales:\n", monthly_sales)
```

```python
bins = [0, 18, 30, 45, 60, 100]
labels = ['Teen', 'Young Adult', 'Adult', 'Middle Age', 'Senior']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
age_group_sales = df.groupby('Age Group')['Total Amount'].sum()
plt.figure(figsize=(6,4))
age_group_sales.plot(kind='bar', color='purple')
plt.title('Total Sales by Age Group')
plt.ylabel('Total Amount')
plt.show()
```

```python
top_products = df.groupby('Product Category')['Quantity'].sum()
plt.figure(figsize=(6,4))
top_products.plot(kind='bar', color='pink')
plt.title('Top Products by Quantity Sold')
plt.ylabel('Total Units Sold')
plt.show()
```

```python
sns.histplot(df['Age'], bins=10, kde=False)
plt.title('Customer Age Distribution')
plt.show()
```

```python
top_customers=df.groupby('Customer ID')['Total Amount'].sum().sort_values(ascending=False).head(10)
print(top_customers)
```

```python
sns.heatmap(df[['Quantity','Price per Unit','Total Amount','Age']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("menu.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
category_counts = df['Category'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(category_counts,  labels=category_counts.index, autopct='%1.1f%%', startangle=170)
plt.title('Percentage of Items per Category')
plt.show()
```

```python

plt.figure(figsize=(8,6))
sns.scatterplot(x='Total Fat', y='Calories', data=df, hue='Category', palette='Set3')
plt.title('Calories vs Total Fat')
plt.xlabel('Total Fat')
plt.ylabel('Calories')
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(x='Sugars', y='Calories', data=df, hue='Category', palette='Set1')
plt.title('Calories vs Sugars')
plt.xlabel('Sugars')
plt.ylabel('Calories')
plt.show()
```

```python
top_items = df.nlargest(5, 'Calories')

top_items = top_items.sort_values('Calories', ascending=False)

# 3️⃣ Display just to confirm
print(top_items[['Item', 'Calories']])

# 5️⃣ Plot 2 — Stacked bar chart for nutrients (% Daily Value)
top_items.set_index('Item')[[
    'Total Fat (% Daily Value)',
    'Saturated Fat (% Daily Value)',
    'Cholesterol (% Daily Value)',
    'Sodium (% Daily Value)',
    'Carbohydrates (% Daily Value)',
    'Dietary Fiber (% Daily Value)',
    'Vitamin A (% Daily Value)',
    'Vitamin C (% Daily Value)',
    'Calcium (% Daily Value)',
    'Iron (% Daily Value)'
]].plot(kind='bar', stacked=True, figsize=(12,6), colormap='tab10')

plt.title('% Daily Value of Nutrients for Top Calorie Items')
plt.ylabel('% Daily Value')
plt.xticks(rotation=45)
plt.legend(title='Nutrient', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
df['Serving_Size_gml'] = df['Serving_Size'].str.extract(r'(\d+)').astype(float)


df['Serving_Unit'] = df['Serving_Size'].str.extract(r'([a-zA-Z]+)')
df.columns = df.columns.str.replace('[()/% ]', '_', regex=True)

print(df.columns)  # ✅ check your cleaned column names
plt.figure(figsize=(8,5))
plt.hist(df['Serving_Size_gml'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Serving Size')
plt.xlabel('Serving Size (grams/ml)')
plt.ylabel('Number of Items')
plt.show()
```

```python
# Top 10 items by Calories
top10_calories = df.nlargest(10, 'Calories')

plt.figure(figsize=(10,6))
sns.barplot(x='Calories', y='Item', data=top10_calories, color='salmon')
plt.title('Top 10 Highest-Calorie Items')
plt.xlabel('Calories')
plt.ylabel('Food Item')
plt.show()

# Top 10 items by Protein
top10_protein = df.nlargest(10, 'Protein')

plt.figure(figsize=(10,6))
sns.barplot(x='Protein', y='Item', data=top10_protein, color='green')
plt.title('Top 10 Protein-Rich Items')
plt.xlabel('Protein (g)')
plt.ylabel('Food Item')
plt.show()

```

```python
print(df.columns.tolist())
nutrients = ['Calories', 'Protein', 'Sodium', 'Sugars']

for nutrient in nutrients:
    plt.figure(figsize=(9,5))
    sns.barplot(x=nutrient, y='Item', data=df.nlargest(10, nutrient), palette='viridis')
    plt.title(f'Top 10 Items by {nutrient}')
    plt.xlabel(nutrient)
    plt.ylabel('Food Item')
    plt.show()
```

```python
numeric_cols = ['Calories', 'Total_Fat', 'Saturated_Fat', 'Cholesterol', 'Sodium',
                'Carbohydrates', 'Dietary_Fiber', 'Sugars', 'Protein']

plt.figure(figsize=(10,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Nutrients')
plt.show()
```

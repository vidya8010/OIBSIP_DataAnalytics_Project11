```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
apps_df=pd.read_csv('apps.csv')
```

```python
user_view_df=pd.read_csv('user_reviews.csv')
```

```python
print("Apps Data:")
print(apps_df.head())
```

```python
print("\n User Reviews Data:")
print(user_view_df.head())
```

```python
apps_df.info()
```

```python
user_view_df.info()
```

```python
apps_df.isnull().sum()
```

```python
merged_df = pd.merge(apps_df, user_view_df, on='App', how='inner')
```

```python
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df.dropna(subset=['Sentiment', 'Translated_Review'], inplace=True)
```

```python
merged_df.info()
```

```python

merged_df['Category'].value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 App Categories on Play Store")
plt.xlabel("Category")
plt.ylabel("Number of Apps")
plt.show()
```

```python
merged_df['Sentiment'].value_counts().plot(kind='bar', color=['green','red','grey'])
plt.title("User Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

```python
avg_rating = merged_df.groupby('Category')['Rating'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=avg_rating.values, y=avg_rating.index)
plt.title("Average Rating per Category")
plt.xlabel("Average Rating")
plt.ylabel("Category")
plt.show()
```

```python
merged_df['Installs'] = merged_df['Installs'].str.replace('[+,]', '', regex=True).astype(int)

sns.scatterplot(data=merged_df, x='Installs', y='Rating', alpha=0.5)
plt.title("Rating vs Installs")
plt.xscale('log')
plt.show()
```

```python
sns.countplot(data=merged_df, x='Type', hue='Sentiment')
plt.title("Sentiment Comparison: Free vs Paid Apps")
plt.show()
```

```python
merged_df.to_csv("cleaned_playstore_data.csv", index=False)
```
